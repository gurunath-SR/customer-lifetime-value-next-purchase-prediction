```python
# ==========================================
# CUSTOMER LIFETIME VALUE PROJECT
# PHASE 2: DATA CLEANING
# ==========================================

# Import Libraries
import pandas as pd
import numpy as np

# Load Dataset
df = pd.read_excel("../data/Online Retail.xlsx")

print("=" * 60)
print("ORIGINAL DATASET")
print("=" * 60)

print("Shape:", df.shape)

# ------------------------------------------
# Create Revenue Column
# ------------------------------------------

df["Revenue"] = df["Quantity"] * df["UnitPrice"]

# ------------------------------------------
# Step 1: Remove Missing Customer IDs
# ------------------------------------------

print("\nMissing Customer IDs:")
print(df["CustomerID"].isnull().sum())

df = df.dropna(subset=["CustomerID"])

print("\nAfter Removing Missing Customer IDs:")
print(df.shape)

# ------------------------------------------
# Step 2: Remove Duplicate Rows
# ------------------------------------------

duplicates = df.duplicated().sum()

print("\nDuplicate Rows:", duplicates)

df = df.drop_duplicates()

print("After Removing Duplicates:")
print(df.shape)

# ------------------------------------------
# Step 3: Remove Cancelled Orders
# InvoiceNo starts with 'C'
# ------------------------------------------

cancelled_orders = df[
    df["InvoiceNo"].astype(str).str.startswith("C")
]

print("\nCancelled Orders:")
print(cancelled_orders.shape[0])

df = df[
    ~df["InvoiceNo"].astype(str).str.startswith("C")
]

print("After Removing Cancelled Orders:")
print(df.shape)

# ------------------------------------------
# Step 4: Remove Negative Quantities
# Returns / Refunds
# ------------------------------------------

negative_quantity = (df["Quantity"] <= 0).sum()

print("\nNegative Quantity Records:")
print(negative_quantity)

df = df[df["Quantity"] > 0]

print("After Removing Negative Quantities:")
print(df.shape)

# ------------------------------------------
# Step 5: Remove Invalid Prices
# ------------------------------------------

invalid_price = (df["UnitPrice"] <= 0).sum()

print("\nInvalid Prices:")
print(invalid_price)

df = df[df["UnitPrice"] > 0]

print("After Removing Invalid Prices:")
print(df.shape)

# ------------------------------------------
# Final Revenue Calculation
# ------------------------------------------

df["Revenue"] = df["Quantity"] * df["UnitPrice"]

# ------------------------------------------
# Final Summary
# ------------------------------------------

print("\n" + "=" * 60)
print("CLEANED DATASET SUMMARY")
print("=" * 60)

print("Final Shape:", df.shape)

print("\nRemaining Missing Values:")
print(df.isnull().sum())

print("\nTotal Revenue:")
print(round(df["Revenue"].sum(), 2))

print("\nUnique Customers:")
print(df["CustomerID"].nunique())

print("\nUnique Products:")
print(df["StockCode"].nunique())

print("\nTotal Transactions:")
print(df["InvoiceNo"].nunique())

# ------------------------------------------
# Save Clean Dataset
# ------------------------------------------

df.to_csv(
    "../data/cleaned_online_retail.csv",
    index=False
)

print("\nCleaned Dataset Saved Successfully!")
```
